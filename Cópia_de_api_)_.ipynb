```python
import os
import shutil
from google.colab import drive

# 1. Montar o Google Drive
print("Montando o Google Drive...")
drive.mount('/content/drive')
print("Google Drive montado com sucesso.")

# 2. Definir os caminhos das pastas
# Caminho da pasta de origem no seu Google Drive
source_folder = '/content/drive/MyDrive/comex stat arquivos'
# Caminho da pasta de destino no ambiente Colab
destination_folder = '/content'

# Certifique-se de que a pasta de destino existe
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)
    print(f"Pasta de destino '{destination_folder}' criada.")

# 3. Copiar os arquivos
print(f"Copiando arquivos de '{source_folder}' para '{destination_folder}'...")

try:
    # Lista todos os arquivos na pasta de origem
    for filename in os.listdir(source_folder):
        source_path = os.path.join(source_folder, filename)
        destination_path = os.path.join(destination_folder, filename)

        # Copia o arquivo, se for um arquivo e não uma pasta
        if os.path.isfile(source_path):
            shutil.copy2(source_path, destination_path)
            print(f"Arquivo copiado: {filename}")
    print("Todos os arquivos foram copiados com sucesso!")
except FileNotFoundError:
    print("Erro: A pasta de origem não foi encontrada. Verifique se o nome da pasta está correto.")
except Exception as e:
    print(f"Ocorreu um erro: {e}")
```

```python
import pandas as pd

# Load the dataframes
try:
    juncao_exp_Bra_v1 = pd.read_csv('/content/juncao_exp_Bra_v1.csv')
    pais_sig = pd.read_csv('/content/PAIS_SIG.csv')
except FileNotFoundError as e:
    print(f"Error loading file: {e}. Please make sure the files are in the correct location.")
    # Exit if files are not found
    exit()
except Exception as e:
    print(f"An unexpected error occurred while loading files: {e}")
    # Exit on other loading errors
    exit()


# Before merging, let's check the column names and data types for the merge key
print("Columns in juncao_exp_Bra_v1:", juncao_exp_Bra_v1.columns)
print("Columns in pais_sig:", pais_sig.columns)
print("\nData type of 'CO_PAIS' in juncao_exp_Bra_v1:", juncao_exp_Bra_v1['CO_PAIS'].dtype)
print("Data type of 'CO_PAIS' in pais_sig:", pais_sig['CO_PAIS'].dtype)

# Ensure the merge key columns have the same data type
try:
    juncao_exp_Bra_v1['CO_PAIS'] = juncao_exp_Bra_v1['CO_PAIS'].astype(int)
    pais_sig['CO_PAIS'] = pais_sig['CO_PAIS'].astype(int)
except ValueError:
    print("Could not convert 'CO_PAIS' to integer in one or both dataframes. Check for non-numeric values.")


# Merge the dataframes based on the country code (CO_PAIS)
# We'll use a left merge to keep all records from the export data
juncao_exp_Bra_v2 = pd.merge(
    juncao_exp_Bra_v1,
    pais_sig,
    on='CO_PAIS',
    how='left'
)

# The user wants to keep both CO_PAIS and NO_PAIS from the merge.
# The merge operation already keeps CO_PAIS and adds NO_PAIS.
# We just need to reorder the columns to place NO_PAIS next to CO_PAIS

# Get the list of columns in the merged dataframe
current_cols = juncao_exp_Bra_v2.columns.tolist()

# Find the index of the 'CO_PAIS' column
try:
    copais_index = current_cols.index('CO_PAIS')

    # Find the index of the 'NO_PAIS' column
    nopais_index = current_cols.index('NO_PAIS')

    # Remove 'NO_PAIS' from its current position
    nopais_col = current_cols.pop(nopais_index)

    # Insert 'NO_PAIS' right after 'CO_PAIS'
    current_cols.insert(copais_index + 1, nopais_col)

    # Reindex the dataframe with the new column order
    juncao_exp_Bra_v2 = juncao_exp_Bra_v2[current_cols]

except ValueError:
    print("Warning: Could not find 'CO_PAIS' or 'NO_PAIS' for reordering.")
    # If columns are not found, keep the columns as they are after the merge


# Display the first few rows of the new DataFrame with country names
print("\nCombined DataFrame (juncao_exp_Bra_v2) with country names:")
display(juncao_exp_Bra_v2.head())

# Save the new DataFrame to a CSV file
output_file_v2 = '/content/juncao_exp_Bra_v2.csv'
juncao_exp_Bra_v2.to_csv(output_file_v2, index=False)
print(f"\nCombined data with country names saved to {output_file_v2}")
```
